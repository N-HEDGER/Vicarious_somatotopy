Load in myelin

```python
surfinfo_wildcard= 'hcp_{sub}/surface-info/{param}.npz'
param='myelin'
subject='999999'

pcx_dir = cortex.db.filestore
def get_surf_info(param: str):
    """gets a surface parameter for a given subject.

    Args:
        param (str): The parameter to get.

    Returns:
        param (np.ndarray): A numpy array of the surface parameter.
    """
    surfinfo = np.load(os.path.join(pcx_dir, surfinfo_wildcard.format(
        sub=subject, param=param)), allow_pickle=True)
    param = np.concatenate([surfinfo['left'], surfinfo['right']])
    return param
```

```python
myelin=get_surf_info(param)
```

Plot out left and right hemisphere.

```python
plotviews=load_plot_pkg_yaml()['plotviews']

mp=Plot(**plotviews['face_area_L'])

# Generate the plot using d as the data for the first plot and d as the data for the second plot and store the result in F1_e
F1_e=mp.uber_plot(dat=myelin,dat2=np.ones_like(myelin))
```

```python
plotviews=load_plot_pkg_yaml()['plotviews']

mp=Plot(**plotviews['face_area_R'])

# Generate the plot using d as the data for the first plot and d as the data for the second plot and store the result in F1_e
F1_e=mp.uber_plot(dat=myelin,dat2=np.ones_like(myelin))
```
